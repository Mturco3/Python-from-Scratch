## 0) Import Libraries and Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

housing_data = pd.read_csv('housing.csv')
```

## 1) Data Overview

```python
print(f"The dataset contains {housing_data.shape[0]} features and {housing_data.shape[1]} attributes.\n")
print(f"The dataset has a total of {housing_data.isnull().sum().sum()} missing values.\n")

for column in housing_data.columns:
    print(f"Column '{column}' has data type: {housing_data[column].dtype}, {housing_data[column].isnull().sum()} missing values and {housing_data[column].nunique()} unique values.")

print(f"\n{housing_data['ocean_proximity'].value_counts()}")
```

```python
print(housing_data.describe())
```

## 2) Data Visualization

```python
housing_data.hist(bins=50, figsize=(20,15))
plt.tight_layout()
plt.show()
```

```python
housing_data.plot(kind='scatter', x='longitude', y='latitude', alpha=0.4,
                  s=housing_data['population']/100, label='population', figsize=(10,7),
                  c='median_house_value', colorbar=True)
plt.legend()
plt.tight_layout()
plt.show()
```
